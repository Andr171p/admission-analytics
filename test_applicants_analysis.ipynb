### Все непоступившие абитуриенты ТИУ

```python
tyuiu_applicants_df = pd.read_csv("ТИУ_непоступившие_абитуриенты_2025.csv")
```

```python
tyuiu_applicants_ids = tyuiu_applicants_df["Id поступающего"].unique()
tyuiu_applicants_ids.tolist()
```

```python
filtered_applicants_df = applicants_df[applicants_df["id"].isin(tyuiu_applicants_ids)]
```

```python
filtered_applicants_df.shape
```

```python
filtered_applicants_df["university_id"].value_counts()
```

```python
uni118_applicants_df = filtered_applicants_df[filtered_applicants_df["university_id"] == 118]
uni118_applicants_df.shape
```

```python
uni118_applicants_df["total_points"].mean()
```

```python
uni118_applicants_df["total_places"] = (
    uni118_applicants_df["reception"]
    .apply(lambda s: int(s.split(",")[1].split(" ")[1]))
)
uni118_applicants_df["total_places"]
```

```python
uni118_applicants_df
```

```python
uni118_applicants_df["is_entered"] = (
    (uni118_applicants_df["place"] < uni118_applicants_df["total_places"]) &
    (uni118_applicants_df["priority"] == 1)
)
```

```python
uni118_applicants_df["is_entered"].value_counts()
```

```python
entered_applicants_df = uni118_applicants_df[uni118_applicants_df["is_entered"] == 1]

entered_applicants_df
```
